# Informações dos atributos

  De Entrada:
1. **fixed acidity (Ácidez Fixa):** Os ácidos orgânicos da uva (tartárico, málico e cítrico) e os originados da fermentação alcoólica (succínico, lático, etc.), que formam a acidez fixa do vinho.  
2. **volatile acidity (Ácidez Volátil):** Corresponde à soma dos ácidos graxos da série acética presentes no vinho no estado livre ou salificado.
3. **citric acid (Ácido Cítrico):** Presente nas uvas em baixa quantidade, nos vinhos o ácido cítrico tem pouca ou nenhuma presença. Nos vinhos tintos desaparece devido à ação de bactérias láticas. Sensorialmente é fresco, porém em alguns casos pode apresentar um leve final amargo.
4. **residual sugar (Açúcar residual):** É a sobra dos açúcares do mosto que não foram convertidos em álcool pelas leveduras durante a fermentação, que dão a doçura do vinho. 
5. **chlorides (Cloretos):** Os vinhos obtidos de vinhedos situados mais próximos do mar apresentam teores mais elevados, devido a irrigação com água salina, ou que possuem solos salobro.
* **sulfur dioxide (Dióxido de enxofre):** Seu uso impacta na fermentação, intensidade das cores do vinhos vermelho e no aroma.
6. **free sulfur dioxide (Dióxido de enxofre livre)** 
7. **total sulfur dioxide (Dióxido de enxofre total):** Corresponde à soma do dióxido de enxofre livre mais o combinado existente no vinho.
8. **density (Densidade):** Quociente entre a massa e o volume desse corpo. Ela pode variar de acordo com a densidade da água e o teor percentual de álcool e açúcar.
9. **pH:** Medida utilizada para quantificar o grau de acidez, lembrando que os ácidos são os principais atributos que contribuem para o que chamamos de sabor azedo do vinho. 
10. **sulphates (Sulfatos):** O teor de sulfatos aumenta progressivamente durante a conservação do vinho.
11. **alcohol (Concentração alcoólica)** Resultado da conversão do açúcar, contido no mosto, em etanol pelas leveduras. 
<br>
De Saída:   
12. **quality (Qualidade)**: Valores numa escala de 0 a 10 que determinam seu grau de excêlencia.

```
import six
import sys
sys.modules['sklearn.externals.six'] = six
from lazypredict.Supervised import LazyClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Carregar o dataset
data = winewhite_df.copy(deep=True)

#OverSampling
X_resampled, y_resampled = SMOTE(random_state=1,k_neighbors=4).fit_resample(data.drop('qualidade',axis=1),data['qualidade'])

# Organizar nossos dados
features = X_resampled
label = y_resampled

A_train, A_test, z_train, z_test = train_test_split(features, label, test_size=0.3, random_state=1)

lazy = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = lazy.fit(A_train, A_test, z_train, z_test)

models
```

```
#Colocar o rótulo nas colunas novamente.
cols= data.columns.drop("qualidade")
bdf = pd.DataFrame(features,columns=cols)
```

```
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import KFold,cross_validate
from sklearn.metrics import classification_report,plot_confusion_matrix
from yellowbrick.model_selection import FeatureImportances,learning_curve

features = bdf

#Classificador Escolhido
clf = ExtraTreesClassifier(n_estimators=100, random_state=1)

x_train, x_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size=0.3,
                                                    random_state=1)
clf.fit(x_train, y_train)
result = clf.score(x_test, y_test)
print("Accuracy: %.2f%%" % (result*100.0))
print()
print(classification_report(y_test,clf.predict(x_test)))

#kfold
kfold = KFold(n_splits=10, random_state=1, shuffle=True)
metricas = ['accuracy', 'f1_weighted', 'precision_weighted' , 'recall_weighted','roc_auc_ovr_weighted']
results_kfold = cross_validate(clf, features, label, cv=kfold, scoring=metricas, verbose=0)
for met in sorted(results_kfold.keys()):
  print(met + " KFold: %.2f%%" % (results_kfold[met].mean()*100.0))
```

```
viz = FeatureImportances(clf)
viz.fit(x_train, y_train)
viz.show();
```

```
learning_curve(clf, features, label,cv=kfold);
```
